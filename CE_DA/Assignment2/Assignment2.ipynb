```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv("Ex1-Class-train.csv")
df
```

```python
df.info()
```

```python
import statistics

df["y"].fillna(statistics.mode(df["y"]), inplace=True)
df["x3"].fillna(statistics.mode(df["x3"]), inplace=True)
df["x4"].fillna(statistics.mode(df["x4"]), inplace=True)
df["x5"].fillna(statistics.mode(df["x5"]), inplace=True)
```

```python
df.info()
```

```python
%matplotlib widget
```

```python
df.dropna(inplace=True)
df
```

```python
df.reset_index(drop=True, inplace=True) 
df
```

```python
# input
x = df.iloc[:, 1:-1]
# output
y = df.iloc[:, -1]
```

```python
from sklearn.preprocessing import OrdinalEncoder

enc = OrdinalEncoder()
enc.fit(x[["x3", "x4", "x5"]])
```

```python
enc.categories_
```

```python
arr = pd.DataFrame(enc.transform(x[["x3", "x4", "x5"]]), columns=["x3", "x4", "x5"])
arr
```

```python
x[["x3", "x4", "x5"]] = arr
x["x4"] = x["x4"] + 5
x["x5"] = x["x5"] + 10
x
```

```python
x = x.drop(["x3","x4","x5"], axis=1)
```

```python
x
```

```python
y
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.25)

print(f"Train data shape of x = {x_train.shape} and y = {y_train.shape}")
print(f"Test data shape of x = {x_test.shape} and y = {y_test.shape}")
```

**DECISION TREE**

```python
from sklearn import tree
from sklearn.tree import DecisionTreeClassifier

tree_model = DecisionTreeClassifier()
tree_model.fit(x_train, y_train)
```

```python
print(tree.export_text(tree_model))
```

```python
y_pred = tree_model.predict(x_test)
y_pred
```

```python
from sklearn.metrics import accuracy_score

acc = accuracy_score(y_test, y_pred)
acc
```

```python
from sklearn.metrics import precision_recall_fscore_support

prf = precision_recall_fscore_support(y_test, y_pred, average = "weighted")
prf
```

**RANDOM FORREST**

```python
from sklearn.ensemble import RandomForestClassifier

rf_model = RandomForestClassifier(n_estimators=288, max_depth=17)
rf_model.fit(x_train, y_train)
```

```python
y_pred = rf_model.predict(x_test)
y_pred
```

```python
acc = accuracy_score(y_test, y_pred)
acc
```

```python
prf = precision_recall_fscore_support(y_test, y_pred, average = "weighted")
prf
```

```python
from sklearn.model_selection import KFold

kf = KFold(n_splits = 5)
acc_list = []

for train_index, test_index in kf.split(x):
  X_train, X_test = x.loc[train_index], x.loc[test_index]
  y_train, y_test = y.loc[train_index], y.loc[test_index]

  model = RandomForestClassifier(n_estimators=288, max_depth=17)
  model.fit(X_train, y_train)

  y_pred = model.predict(X_test)
  acc = accuracy_score(y_test, y_pred)
  acc_list.append(acc)

print("------------------------------------")
print("Random Forest")
print(" Mean Accuracy :", np.mean(acc_list))
```
